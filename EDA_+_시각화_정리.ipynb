### 시각화

```python
# 라이브러리 로딩 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
# 그림을 바로 볼 수 있게 해줌
%matplotlib inline
# 그래프 해상도를 높여줌
%config InlineBackend.figure_format = 'retina'

import pandas as pd
import numpy as np
```

### Line chart

```python
year = [1950,1970,1990,2010]
pop = [2.519,3.692,5.263,6.972]

plt.plot(year, pop)
plt.title('Matplotlib 1')
plt.show()
```

```python
# 객체지향식
fig,ax = plt.subplots(nrows = 1,ncols = 1)
ax.set_title("Matplotlib 2")
ax.plot(year,pop)
```

```python
# pandas 그래프
dic = dict(year = year,pop = pop)
df = pd.DataFrame(dic)
df.plot(x = 'year',y ='pop',kind='line')
plt.title("Matplotlib 3")
```

### Scatter plot

```python
plt.scatter(year, pop)
plt.show()
```

```python
df.plot(y = 'year',x = 'pop', kind = 'scatter')
```

```python
# alpha : 투명도 조절 (연하게 0 진하게 1)
df.plot(y = 'year',x = 'pop',kind='scatter',alpha = 0.4)
```

```python
# c에 따라서 그라데이션 조절가능
df.plot(y = 'year',x = 'pop', kind='scatter',c = 'pop', colormap = 'Blues')
```

### Histogram

```python
values = [0,1.6,1.4,2.2,2.5,2.6,3.2,3.5,3,3.9,4.2,6]
```

```python
plt.hist(values, bins = 6, color = 'red')
plt.show()
```

### Borplot

```python
objects = ['Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp']
performance = [10,8,6,4,2,1]
```

```python
plt.bar(objects, performance)
plt.show()
```

```python
df.groupby('col1').mean().drop('liters',axis = 1).plot(kind='bar')
df.groupby('col1').mean().drop('liters',axis = 1).plot(kind='bar', stacked = True) # 쌓기
```

### Pie Chart

```python
plt.pie(performance, labels = objects)
plt.show()
```

### Customize
#### Axis, Title

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

plt.figure(figsize=(10,3))
plt.plot(year, pop)

plt.xlabel('year')
plt.ylabel('pop')

plt.title('World Population Projections')
plt.show()
```

#### ticks

```python
plt.plot(year, pop)
plt.yticks([0,2,4,8,10],['a','b','c','d','e'])
plt.show()
```

#### Common Axis

```python
year = [1950, 1970, 1990, 2010]
pop1 = [2.519, 3.692, 5.263, 6.972]
pop2 = [1.231, 2.252, 2.988, 4.334]

plt.plot(year, pop1, 'red')
plt.plot(year, pop2, 'blue')
plt.show()
```
